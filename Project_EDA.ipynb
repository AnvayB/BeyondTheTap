```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'water_data.csv' with your file path)
df = pd.read_csv('combined_water_quality_pipeline_data.csv')

# Display basic information and first few rows
print("Dataset Info:")
df.info()
print("\nFirst 5 rows:")
print(df.head())

# Identify categorical and numerical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

print("\nCategorical Columns:", categorical_cols)
print("\nNumerical Columns:", numerical_cols)

# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(df[numerical_cols].describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Visualizing missing values using heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values Heatmap')
plt.show()

# Univariate Analysis: Distribution of numeric features
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()

# Boxplot for outlier detection
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Correlation heatmap (only numerical columns)
plt.figure(figsize=(12, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()

# Pairplot for numerical features
sns.pairplot(df[numerical_cols])
plt.show()

print("EDA Completed Successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Define the target column (Change as per your dataset)
target_column = 'Target'  # Change this to your actual target column
features = df.select_dtypes(include=np.number).drop(columns=[target_column])

# Handle missing values
features = features.fillna(features.median())

# Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(features, df[target_column], test_size=0.2, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN model
k = 5  # Adjust for optimization
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train_scaled, y_train)

# Predictions
y_pred = knn.predict(X_test_scaled)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'KNN Model Accuracy: {accuracy:.2f}')
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Show some sample predictions
sample_df = X_test.copy()
sample_df['Actual'] = y_test.values
sample_df['Predicted'] = y_pred
print("\nSample Predictions:")
print(sample_df.head())

# Confusion Matrix Plot
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='d')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix')
plt.show()

# Distribution of predicted vs actual
plt.figure(figsize=(6,4))
sns.histplot(y_test, color='blue', label='Actual', kde=True)
sns.histplot(y_pred, color='red', label='Predicted', kde=True, alpha=0.6)
plt.legend()
plt.title('Distribution of Actual vs Predicted')
plt.show()

```
